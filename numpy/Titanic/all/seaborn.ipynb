```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
        
sinplot()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)#implot
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',hue='time',data=tips)#orient='h'
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,split=True)
```

```python
g = sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

```python
g=sns.FacetGrid(tips,col='sex',hue='smoker')
g.map(plt.scatter,'total_bill','tip',alpha=.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips,col='day',height=3,aspect=.6)
g.map=(sns.barplot,'sex','total_bill')
```

```python
g=sns.FacetGrid(tips,hue='sex',palette='Set1',size=5,hue_kws={'marker':["o","v"]})
g.map(plt.scatter,'total_bill','tip',s=50,alpha=.7,linewidth=.5,edgecolor='white')
g.add_legend()
```

```python
with sns.axes_style('white'):
    g=sns.FacetGrid(tips,row='sex',col='smoker',margin_titles=True,size=2.5)
g.map(plt.scatter,'total_bill','tip',color='#334477',edgecolor='white',lw=.5)
g.set_axis_labels('total bill','tip')
g.set(xticks=[10,30,50],yticks=[2,6,10])
g.fig.subplots_adjust(wspace=.02,hspace=.02)#
```
