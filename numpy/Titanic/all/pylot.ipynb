```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
titanic_survival = pd.read_csv('train.csv')
```

```python
new_titanic_survival = titanic_survival.dropna(subset=['Age'])
pclass = new_titanic_survival['Pclass']
#print(pclass)
class_notone = pclass[pclass != 1]
#print(class_notone.index)
new_titanic_survival=new_titanic_survival.drop(class_notone.index)
#print(new_titanic_survival)
```

```python
titanic_reindexed = new_titanic_survival.reset_index(drop=True)
```

```python
data_showed = titanic_reindexed[0:10]
plt.plot(data_showed.index,data_showed.Age)
plt.xlabel('index')
plt.xticks(rotation=45)
plt.ylabel('age')
plt.title('something xiaJB write de')
plt.show()
```

```python
bar_heights=data_showed.Age.values
bar_positions=np.arange(10)+0.75
fig,ax=plt.subplots()
#ax.bar(bar_positions,bar_heights,0.5)
ax.barh(bar_positions,bar_heights,0.5)
ax.set_xticklabels(data_showed.Name,rotation=45)
ax.set_xlabel('Name')
ax.set_ylabel('Age')
ax.set_title('boring')
plt.show()
```
